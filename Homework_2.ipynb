## Задание 3

```python
from math import sin
```

```python
x=np.linspace(-3,3,51)
x
```

```python
y1=x**2
y2=2*x+0.5
y3=3*x-1.5
y4=np.array([sin(i) for i in x])
```

```python
fig, ax = plt.subplots(nrows=2, ncols=2)
fig.subplots_adjust(wspace=0.3,hspace=0.4)
fig.set_size_inches(8, 6)
ax[0,0].set_xlim([-5,5])
ax[0,0].plot(x,y1)
ax[0,0].set_title('График y1')
ax[0,1].plot(x,y2)
ax[0,1].set_title('График y2')
ax[1,0].plot(x,y3)
ax[1,0].set_title('График y3')
ax[1,1].plot(x,y4)
ax[1,1].set_title('График y4')
```

```python
f.plot(kind='bar',logy=True)
plt.show()
```

```python
y1=df[df['Class']==1]['V1'].values
y2=df[df['Class']==0]['V1'].values
```

```python
fig, ax = plt.subplots(nrows=1, ncols=2)
fig.set_size_inches(8, 6)
fig.subplots_adjust(hspace=0.5)
ax[0].hist(y1,bins=20,alpha=0.5,color='Red',density=True)
ax[0].set_title('Class 1')
ax[1].hist(y2,bins=20,alpha=0.5,color='Grey',density=True)
ax[1].set_title('Class 0')
```
